```python
def generate_score(positions, occupied):
    controlled_tiles = {}
    for player in positions.keys():
        controlled_tiles[player] = None
    
    
    for i in range(30):
        for j in range(20):
            if (i,j) not in occupied:
                tile = (i,j)
                num_players = len(positions)
                closest_player = -1
                min_cost = 10000000
                for player in range(len(num_players)):
                    player_cost = a_star(occupied, tile, positions[player])
                    if player_cost < min_cost:
                        closest_player = player
                        min_cost = player_cost
    
        controlled_tiles[closest_player] = tile

    my_tiles = len(controlled_tiles[MY_ID])

    num_opponent_tiles = 0

    for player in positions.keys():
        if i != MY_ID:
            num_opponent_tiles += len(positions[player])

    my_mult = 1000000
    opp_mult = 100000

    return my_tiles * my_mult + num_opponent_tiles * opp_mult
```

```python
def a_star(occupied, start, end):
    manhattan = abs(start[0] - end[0]) + abs(start[1] - end[1])

    return manhattan
```
